## Matplotlib Plot creation workflow

```python
# 0. import the matplotlib

%matplotlib inline
import matplotlib.pyplot as plt

# 1. Preparing Data

x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot

fig, ax = plt.subplots(figsize=(10,10)) # (width, height)

# 3. Plot Data

ax.plot(x, y)

# 4. Customize plot

ax.set(title = "Simple Plot",
      xlabel = "X-Axis",
      ylabel = "Y-Axis")

# 5. Optional Step to save the figure

fig.savefig("first_plot.png")
```

# Plotting Bar Graph using Dictionary

```python
cookies = {"Almond Butter Cookies": 10,
          "Peanut Butter Cookies": 8,
          "Cashew Butter Cookies": 12}
cookies
```

```python
fig, ax = plt.subplots()
ax.bar(cookies.keys(), height=cookies.values()) # Positional Arguments shit. Just add the second argument.
ax.set(title = "Bar Graph",
      xlabel = "Cookies",
      ylabel = "Prices in $");
```

```python
fig, ax = plt.subplots()
ax.barh(list(cookies.keys()), list(cookies.values())) # Need to convert the values into lists as originally they are dictionary keys
ax.set(title = "Horizontal Bar",
      xlabel = "Cookies",
      ylabel = "Prices in $");
```

```python
z = np.random.randn(100) # randn = Normal Distribution
fig, ax = plt.subplots()
ax.hist(z)
ax.set(title = "Histogram",
      xlabel = "X-axis",
      ylabel = "Y-axis");
```

## Ploting using OO Method

```python
# Preparing the data - we will try to get the data for people who are over 50

over_50 = heart_disease[heart_disease["age"] > 50]
over_50
```

```python
# Now we will try to plot a scatter plot 
# with an addition of using "target" column which has values 0 and 1 to be used as a way to define colour

over_50.plot(x="age",
            y="chol",
            c= "target",
            kind="scatter");
```

```python
# The above graph is not that pleasant to eyes
# We will try to beautify it by using OO Method

fig, ax = plt.subplots(figsize=(10,6))

# fig = figure, ax = axes. We are adjusting the size of figure in the above line of code.

over_50.plot(x="age",
            y="chol",
            c="target",
            kind="scatter",
            ax=ax);
```

```python
# We can see that the figure looks better now
# It has got x and y labels too now
# We will try to adjust the x axis range just for an example

fig, ax = plt.subplots(figsize=(10,6))

over_50.plot(x="age",
            y="chol",
            c="target",
            kind="scatter",
            ax=ax);

ax.set_xlim(45,100);
```

```python
# Back to the previous plot we made using pyplot + OO method

fig, ax = plt.subplots(figsize=(10,6))

over_50.plot(x="age",
            y="chol",
            c= "target",
            kind="scatter",
            ax=ax);
```

```python
plt.style.use("seaborn-whitegrid")

fig, ax = plt.subplots(figsize=(10,10),
                      nrows=2,
                      ncols=1,
                      sharex=True)

scatter_plot_1 = ax[0].scatter(x=over_50["age"],
                         y=over_50["chol"],
                         c=over_50["target"],
                         cmap = "spring");

ax[0].set(title = "Heart Disease and Cholesterol Levels",
      xlabel = "Age",
      ylabel = "Cholesterol Levels");

ax[0].legend(*scatter_plot_1.legend_elements(), title="target");

ax[0].axhline(over_50["chol"].mean(),
      linestyle="--");


scatter_plot_2 = ax[1].scatter(x=over_50["age"],
                              y=over_50["thalach"],
                              c=over_50["target"],
                              cmap = "winter");

ax[1].set(title = "Heart Disease and Maximum Heart Rate",
         xlabel = "Age",
         ylabel = "Maximum Heart Rate");

ax[1].axhline(over_50["thalach"].mean(),
            linestyle="--",
             color="green");

ax[1].legend(*scatter_plot_2.legend_elements(), title="target");

fig.suptitle(t = "Heart Disease Analysis", fontsize=16, fontweight="bold");
```

```python
# In the above plots we can see an extra white grid line near the start of y axis and on the top of x axis
# We can remove that by using set_xlim and set_ylim by mentioning the x and y range which we want to see

# Deciding the style of the plot

plt.style.use("seaborn-whitegrid")

# Creation of figure and axes

fig, ax = plt.subplots(figsize=(10,10),
                      nrows=2,
                      ncols=1,
                      sharex=True)

# Creating the scatter plot by defining the values to be plotted

scatter_plot_1 = ax[0].scatter(x=over_50["age"],
                         y=over_50["chol"],
                         c=over_50["target"],
                         cmap = "spring");

# Customizing the plot by giving it labels

ax[0].set(title = "Heart Disease and Cholesterol Levels",
      xlabel = "Age",
      ylabel = "Cholesterol Levels");

# Change xlimit and ylimit

ax[0].set_xlim([50,80])
ax[0].set_ylim([100,600])

# Giving the legends. USe the object used to create the scatter plot.

ax[0].legend(*scatter_plot_1.legend_elements(), title="target");

# For the horizontal mean line in the plot

ax[0].axhline(over_50["chol"].mean(),
      linestyle="--",
      color="green");

# Creating the scatter plot by defining the values to be plotted

scatter_plot_2 = ax[1].scatter(x=over_50["age"],
                              y=over_50["thalach"],
                              c=over_50["target"],
                              cmap = "spring");

# Customizing the plot by giving it labels

ax[1].set(title = "Heart Disease and Maximum Heart Rate",
         xlabel = "Age",
         ylabel = "Maximum Heart Rate");

# For the horizontal mean line in the plot

ax[1].axhline(over_50["thalach"].mean(),
            linestyle="--",
            color="green");

# Change xlimit and ylimit

ax[1].set_xlim([50,80]) # We don't really need to change the xlimit since the x axis is shared in both the plots
ax[1].set_ylim([60,200])

# Giving the legends. USe the object used to create the scatter plot.

ax[1].legend(*scatter_plot_2.legend_elements(), title="target");

# Giving the figure a title and adjusting the font as per requirement

fig.suptitle(t = "Heart Disease Analysis", fontsize=16, fontweight="bold");
```
