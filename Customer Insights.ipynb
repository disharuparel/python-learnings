# Customer Insights

> This notebook has been created to get Customer Insights for the dataset which has the following features:

1. InvoiceNo - Invoice number. Nominal. A 6-digit integral number uniquely assigned to
each transaction. If this code starts with the letter 'c', it indicates a cancellation.
2. StockCode - Product (item) code. Nominal. A 5-digit integral number uniquely
assigned to each distinct product.
3. Description - Product (item) name. Nominal.
4. Quantity - The quantities of each product (item) per transaction. Numeric.
5. InvoiceDate - Invoice date and time. Numeric. The day and time when a transaction
was generated.
6. UnitPrice - Unit price. Numeric. Product price per unit in sterling (Â£).
7. CustomerID- Customer number. Nominal. A 5-digit integral number uniquely
assigned to each customer.
8. Country - Country name. Nominal. The name of the country where a customer
resides.

# Data Analysis

```python
# Let's first create a column which has the product of Quantity and Unit Price 
# so we can get the total amount spent on that product

total = []
total = data_comp["Price"]*data_comp["Quantity"]
total[:10]
data_comp["Expenses"] = total
data_comp.head()
```

```python
# Calculating the amount spent by customers

expenses = dict(data_comp.groupby(by="Customer ID")["Expenses"].sum())
expenses
```

```python
# Sorting the customers based on who has spent the most

most_spent = {k:v for k, v in sorted(expenses.items(), key=lambda item:item[1], reverse=True)}
most_spent
```

```python
# Top 10 customers who have spent the most

most_spent_cust = list(most_spent.keys())
most_spent_cust[:10]
```

```python
# The amount spent by the top 10 customers

most_spent_amount = list(most_spent.values())
most_spent_amount[:10]
```

```python
# Let's take the average spent by all the customers

np.mean(most_spent_amount)
```

```python
# Let's take the median spent by all the customers

np.median(most_spent_amount)
```

As we can see that the mean >> median. It means that there are a lot of outliers which indicate that the purchase of the top customers outweigh the purchase of the rest of the customers. These top customers who purchase a lot can be called our high monetary value customers and these should be retained for us to keep making profits. Let us know try to find a threshold which can help us distinguish between the high monetary value customers and the rest of customers.

```python
# Let's plot a histogram to understand from where we can observe the customers who have spent a lot of money

fig, ax = plt.subplots()
ax.hist(most_spent_amount, bins=2000)
ax.set(title = "Expenses",
      xlabel = "Sterlings",
      ylabel = "Frequency")
ax.set_xlim([-10, 30000])
ax.set_ylim([0,1600]);
```
